```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
data_name='distance_data'
```

```python
data={}
```

```python
for i in range(2,11):
    data[i] = pd.read_csv(data_name+str(i)+'.csv')
```

```python
data
```

```python
data_time = data[2]['time'].unique()
```

```python
def vis_img(data_x,data_y,title,data_name,name):
    #print(data_x[0])
    plt.plot(data_x[0],data_y[0],label = data_name[0])
    plt.plot(data_x[1],data_y[1],label = data_name[1])
    plt.plot(data_x[2],data_y[2],label = data_name[2])
    plt.plot(data_x[3],data_y[3],label = data_name[3])
    plt.plot(data_x[4],data_y[4],label = data_name[4])
    plt.xlabel('k(cluster classes nums)')
    plt.ylabel('distance')
    plt.title(title)
    plt.legend()
    plt.savefig('img/'+name+'.png')
    plt.close()
```

```python
index = np.random.choice(data_time,5, replace=False, p=None)
```

```python
data_x_=[]
data_y_=[]
data_name=[]
for tname in data_time:
    data_name.append(tname)
    data_x=[]
    data_y=[]
    title = 'distance and k'
    for k in range(2,11):
        data_x.append(k)
        value = data[k][data[k]['time']==tname]
        data_y.append(value['distance'].mean())
    data_x_.append(data_x)
    data_y_.append(data_y)
```

```python
for i in range(1,10):
    vis_img(data_x_[(i-1)*5:i*5],data_y_[(i-1)*5:i*5],'Clster Distance',data_name[(i-1)*5:i*5],str(i))
```

```python
def vis_result(data_x,data_y,title,data_name,name):
    #print(data_x[0])
    plt.plot(data_x[0],data_y[0],label = data_name[0])
    plt.plot(data_x[1],data_y[1],label = data_name[1])
    plt.plot(data_x[2],data_y[2],label = data_name[2])
    plt.plot(data_x[3],data_y[3],label = data_name[3])
    plt.plot(data_x[4],data_y[4],label = data_name[4])
    plt.xlabel('k(cluster classes nums)')
    plt.ylabel('%')
    plt.title(title)
    plt.legend()
    plt.savefig('img/result'+name+'.png')
    plt.close()
```

```python
data_result_name='result_'
data={}
for i in range(2,11):
    data[i] = pd.read_csv(data_result_name+str(i)+'.csv')
```

```python
data_x_=[]
data_y_=[]
data_name=[]
evalue_name='accuracy'
for num in range(len(data_time)):
    data_name.append(data_time[num])
    data_x=[]
    data_y=[]
    
    #print(value)
    for k in range(2,11):
        data_x.append(k)
        value = data[k].iloc[num,:]
        data_y.append(value[1])
    data_x_.append(data_x)
    data_y_.append(data_y)
```

```python
for i in range(1,10):
    vis_result(data_x_[(i-1)*5:i*5],data_y_[(i-1)*5:i*5],'Clster Classification Accuracy',data_name[(i-1)*5:i*5],str(i))
```
